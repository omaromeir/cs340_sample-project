```python
import os
import pandas as pd
from supabase import create_client, Client
```

```python
SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
if not SUPABASE_URL or not SUPABASE_KEY:
    raise ValueError("Supabase URL or API key not found. Please set the environment variables.")
```

```python
# Create a client for Supabase
supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)

# Fetch data from a table (e.g., 'students')
response = supabase.table('student').select('*').execute()
```

```python
# Convert the response data to a pandas DataFrame and display it as a table
if response.data:
    df = pd.DataFrame(response.data)
    display(df)  # This will display the DataFrame in a table format in Jupyter Notebook
else:
    print("No data found.")
```
